#Duplicate data

```python
import pandas as pd

# Using the duplicated() method to identify duplicate rows
duplicate_rows = data[data.duplicated()]

# Displaying the duplicate rows
print("Duplicate Rows:")
print(duplicate_rows)
```

#Missing Data

```python
# Checking for missing data
missing_data = data.isnull().sum()

# Printing the number of missing values in each column
print("Missing Data:")
print(missing_data)

# Remove rows with missing data
data = data.dropna()

# Confirm that missing rows are removed
print("Shape after removing missing data:", data.shape)
```

#Data Cleaning

```python
#remoning unneccessary data
data.drop(columns=['url'], inplace=True)
data
```

#Data Formatting

```python
from sklearn.preprocessing import LabelEncoder

# Create a LabelEncoder object
label_encoder = LabelEncoder()

# Fit and transform the 'status' column using label encoding
data['status'] = label_encoder.fit_transform(data['status'])
data
```

```python
# Assuming 'data' is your DataFrame
data.to_csv('cleaned_data.csv', index=False)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd


sns.set_style("whitegrid")

# Reshape the DataFrame into long format
melted_data = pd.melt(data.reset_index(), id_vars='index')

plt.figure(figsize=(15, 8))
sns.lineplot(data=melted_data, x='index', y='value', hue='variable', palette='colorblind')
plt.title('Line Graph of All Variables')
plt.xlabel('Index')
plt.ylabel('Value')
plt.legend(title='Variable', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Set the style of seaborn plots
sns.set_style("whitegrid")

numerical_columns = data.select_dtypes(include=['int64', 'float64']).columns
num_numerical = len(numerical_columns)
num_rows = 3
num_cols = 2

# Plot histograms numerical columns
fig, axes = plt.subplots(num_rows, num_cols, figsize=(15, 15))
for i, column in enumerate(numerical_columns):
    row_index = i // num_cols
    col_index = i % num_cols
    if row_index < num_rows and col_index < num_cols:
        sns.histplot(data[column], bins=30, kde=True, color='skyblue', ax=axes[row_index, col_index])
        axes[row_index, col_index].set_title(f'Histogram of {column}')
        axes[row_index, col_index].set_xlabel(column)
        axes[row_index, col_index].set_ylabel('Frequency')

if num_numerical < num_rows * num_cols:
    for j in range(num_numerical, num_rows * num_cols):
        axes[j // num_cols, j % num_cols].axis('off')

plt.tight_layout()
plt.show()
```
